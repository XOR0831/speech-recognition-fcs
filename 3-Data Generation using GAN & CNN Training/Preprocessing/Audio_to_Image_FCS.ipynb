# Convert Audio to Mel Spectrogram

```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
```

```python
# Path to save and load
DATA_Audio_train = "Data_Audio_V2/Train_new/"
DATA_Audio_test = "Data_Audio_V2/Test_new/"
DATA_Images_train = "Data_Images_V2/Train_new/"
DATA_Images_test = "Data_Images_V2/Test_new/"
```

## Training Data

```python
# Loop for every classes
for classes in os.listdir(DATA_Audio_train):
    i = 0
    # Check if directory exists
    if not os.path.exists(DATA_Images_train + classes):
        # Create directory
        os.makedirs(DATA_Images_train + classes)
    # Loop for every file in every classes
    for file in os.listdir(DATA_Audio_train + classes):
        print("Loading:" + str(file))
        # Load Audio
        samples, s = librosa.load(DATA_Audio_train + classes + "/" + file)
        # Mel Spectrogram Specs
        n_fft = 2048
        hop_length = 512
        n_mels = 128
        sample_rate = 16000
        # Convert to Mel Spectrogram (power)
        mel_spec_power = librosa.feature.melspectrogram(samples, sr=sample_rate, n_fft=n_fft, 
                                                        hop_length=hop_length, 
                                                        n_mels=n_mels)
        # Convert to Mel Spectrogram (decibel)
        mel_spec_db = librosa.power_to_db(mel_spec_power, ref=np.max)
        # Save image
        plt.imsave(DATA_Images_train + classes + "/" + str(file) + '.png', mel_spec_db)
        i+=1
        print("Saved")
```

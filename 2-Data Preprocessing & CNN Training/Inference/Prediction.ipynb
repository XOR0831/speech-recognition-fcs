```python
from squeezenet_softmax import SqueezeNet
```

```python
model_squeezenet = SqueezeNet(input_shape=(32,32,3), classes=5, pooling='max')
```

```python
model_squeezenet.load_weights('../../../Downloads/Trained-SqueezeNet-Model-Cry-Type-V5-12-26-2019-V5-2-Epoch 47--Train Acc 1.00--Val Acc 0.97--Weights.h5')
```

```python
classes_required = ['Eairh', 'Eh', 'Heh', 'Neh', 'Owh']
classes_required
```

```python
import numpy as np
from keras.preprocessing import image
import os
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
testX = []
testY = []
for i in os.listdir("../../../Desktop/School/PD/donateacry-corpus-master/Testing/Cry Types/Data V5/Data_Images/Test/"):
    for files in os.listdir("../../../Desktop/School/PD/donateacry-corpus-master/Testing/Cry Types/Data V5/Data_Images/Test/"+i):
        testX.append("../../../Desktop/School/PD/donateacry-corpus-master/Testing/Cry Types/Data V5/Data_Images/Test/"+i+"/"+files)
        testY.append(classes_required.index(i))
    
print(testX)
print(testY)
```

```python
y = []
img_width, img_height = 32, 32
for i in range(len(testX)):
    print(testX[i])
    img = image.load_img(testX[i], target_size = (img_width, img_height))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis = 0)

    Y_pred = model_squeezenet.predict(img)
    y_pred = np.argmax(Y_pred, axis=1)
    y.append(y_pred)
    print("Predicted: " + str(classes_required[y_pred[0]]) + "  Actual:" + str(classes_required[testY[i]]))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(testY,y,target_names=classes_required))
```

```python
trainX = []
trainY = []
for i in os.listdir("./Data_Images_V2/Train/"):
    for files in os.listdir("./Data_Images_V2/Train/"+i):
        trainX.append("./Data_Images_V2/Train/"+i+"/"+files)
        trainY.append(classes_required.index(i))
    
print(trainX)
print(trainY)
```

```python
print(classification_report(trainY,Y,target_names=classes_required))
```
